```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets

dataset = datasets.load_iris()
inputs = dataset.data
labels = dataset.target

print('input shape:', inputs.shape)
print('label shape:', labels.shape)

#n_componentsで圧縮したい次元を指定してPCAをインスタンス化する
pca = PCA(n_components=2)
#入力データの次元圧縮を行う
transformed_inputs = pca.fit_transform(inputs)
print('input shape after being transformed:', transformed_inputs.shape)

for label in np.unique(labels):
    #それぞれのクラス名を設定
    classes = ['Iris-Setosa', 'Iris-Versicolour', 'Iris-Virginica']
    #それぞれのクラス名に対する色の設定
    colors = ['red', 'blue', 'green']
    #150個の値を含んだリストlabelsの値うち、forループ中の各labelの値と同じ
    #値を持つ場合、transformed_inputsの0次元目(第一主成分)を横軸の値、
    #1次元目(第二主成分)を縦軸のに設定して散布する
    plt.scatter(transformed_inputs[labels == label, 0],
               transformed_inputs[labels == label, 1],
               c=colors[label], label=classes[label])

plt.title('PCA')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```
